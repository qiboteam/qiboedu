```python
import numpy as np
import matplotlib.pyplot as plt
import rich
from scipy.optimize import curve_fit
from qibolab import ConfigKinds, DcChannel, IqChannel, Platform, Qubit, Parameters, PulseSequence, Delay, Sweeper, Delay, AveragingMode, Parameter
from qibolab.instruments.emulator import (
    DriveEmulatorConfig,
    EmulatorController,
    FluxEmulatorConfig,
    HamiltonianConfig,
)
```

## Introduction to Qibolab

```python
config = {
  "settings": {
    "nshots": 1024,
    "relaxation_time": 0
  },
  "configs": {
    "emulator/bounds": {
    "kind": "bounds",
    "waveforms": 1000000,
    "readout": 50,
    "instructions": 200
  },
  "hamiltonian":{
    "transmon_levels": 2,
    "single_qubit": {
      "0": {
        "frequency": 5e9,
        "sweetspot": 0.02,
        "anharmonicity": -200e6,
        "asymmetry": 0.0,
        "t1": {
          "0-1": 1000
        },
        "t2": {
          "0-1": 1900
        }
      }
    },
    "kind": "hamiltonian"
  },
    "0/drive": {
      "kind": "drive-emulator",
      "frequency": 5e9,
      "scale_factor": 0.1591549
    },
    "0/drive12": {
      "kind": "drive-emulator",
      "frequency": 4.8e9,
      "scale_factor": 0.1591549
    },
    "0/flux": {
        "kind": "flux-emulator",
        "offset": 0.02
      },
  "0/probe": {
    "kind": "iq",
    "frequency": 5200000000.0
  },
  "0/acquisition": {
    "kind": "acquisition",
    "delay": 0.0,
    "smearing": 0.0,
    "threshold": 0.0,
    "iq_angle": 0.0,
    "kernel": None
  }
  },
  "native_gates": {
    "single_qubit": {
      "0": {
        "RX": [
          [
            "0/drive",
            {
              "duration": 40,
              "amplitude": 0.1594,
              "envelope": {
                "kind": "gaussian",
                "rel_sigma": 0.2
              },
              "relative_phase": 0.0,
              "kind": "pulse"
            }
          ]
        ],
        "RX90": [
          [
            "0/drive",
            {
              "duration": 40,
              "amplitude": 0.07975,
              "envelope": {
                "kind": "gaussian",
                "rel_sigma": 0.2
              },
              "relative_phase": 0.0,
              "kind": "pulse"
            }
          ]
        ],
        "MZ": [
          [
            "0/acquisition",
            {
              "kind": "readout",
              "acquisition": {
                "kind": "acquisition",
                "duration": 100.0
              },
              "probe": {
                "duration": 100.0,
                "amplitude": 0.1,
                "envelope": {
                  "kind": "gaussian",
                  "rel_sigma": 0.2
                },
                "relative_phase": 0.0,
                "kind": "pulse"
              }
            }
          ]
        ],
        "CP": None
      }
    }
    }
  }
```

```python
ConfigKinds.extend([HamiltonianConfig, DriveEmulatorConfig, FluxEmulatorConfig])

qubits = {}
channels = {}

for q in range(1):
    qubits[q] = qubit = Qubit.default(q)
    channels |= {
        qubit.drive: IqChannel(mixer=None, lo=None),
        qubit.flux: DcChannel(),
    }

# register the instruments
instruments = {
    "dummy": EmulatorController(address="0.0.0.0", channels=channels),
}

platform = Platform(
    name="emulator",
    parameters=Parameters(**config),
    instruments=instruments,
    qubits=qubits,
)
```

### Creating a sequence to measure the qubit

```python
natives = platform.natives.single_qubit[0] # native gates available for qubit 0
# intialize empty pulse sequence
seq = PulseSequence()
# let's append a measurement at the end
# NOTE: With + we append to the pulse sequence (starting from the initial time)
seq += natives.MZ()
```

```python
results_ground = platform.execute([seq], nshots=1000)
readout_id_ground = seq.acquisitions[0][1].id
```

```python
def barplot(res: np.ndarray):
    ax = plt.imshow(res[np.newaxis, :], cmap="binary_r", aspect=200, interpolation="nearest").axes
    ax.tick_params(axis="both", left=False, labelleft=False, bottom=False)
    return ax

barplot(results_ground[readout_id_ground]);
```

```python
results_ground_averaged = platform.execute([seq], averaging_mode = AveragingMode.CYCLIC, nshots=1000)
readout_id = seq.acquisitions[0][1].id
```

```python
print(results_ground_averaged[readout_id])
```

### Observe decoherence by adding a delay before the measurement

```python
results_delay = platform.execute([seq], nshots=1000)
readout_id_delay = seq.acquisitions[0][1].id
```

### Using sweepers in Qibolab: T1 experiment

```python
natives = platform.natives.single_qubit[0]

seq = PulseSequence()

# retrieve acquisition channel
acquisition = platform.qubits[0].acquisition

seq += natives.RX()
delay = Delay(duration=2000)
seq.append((acquisition, delay))
seq += natives.MZ()


# introduce sweeper
delay_range = np.arange(100,3000,100)
sweeper = Sweeper(
    parameter = Parameter.duration,
    pulses = [delay],
    values=delay_range
)
```

```python
rich.print(seq)
```

```python
results_t1 = platform.execute([seq], [[sweeper]], nshots=2000, averaging_mode=AveragingMode.CYCLIC)
readout_id_t1 = seq.acquisitions[0][1].id
p1s_t1 = results_t1[readout_id_t1]
```

```python
def plot(x, y, xlabel="Data", ylabel=r"$P_1$"):
    plt.plot(x, y, label=xlabel)
    plt.xlabel("Delay [ns]")
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()

plot(x=delay_range, y=p1s_t1)
```

```python
def exp_decay(x, *params):
    return params[0] - params[1]*np.exp(-x/params[2])
    
def fit_exp(x, y):
    popt, pcov = curve_fit(exp_decay, x, y, p0=[0.5, 0.5, 1000])
    return popt

popt = fit_exp(x=delay_range, y=p1s_t1)
```

```python
def plot_with_fit(x, y, xlabel="Data", ylabel=r"$P_1$", fun=exp_decay, popt=popt):
    plt.plot(x, y, label=xlabel)
    plt.plot(x, fun(x, *popt), label="Fit")
    plt.xlabel("Delay [ns]")
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()

plot_with_fit(delay_range, p1s_t1, fun=exp_decay, popt=popt)
```

```python
print(f"The estimated T1 is {np.round(popt[2])}")
print(f"The simulation value used for T1 is {platform.config("hamiltonian").single_qubit[0].t1[0,1]}")
```

### Using sweepers in Qibolab: T2 experiment

```python
natives = platform.natives.single_qubit[0]

seq = PulseSequence()

# we need to retrieve both the acquisition and the drive channels
acquisition = platform.qubits[0].acquisition
drive = platform.qubits[0].drive
delays = 2* [Delay(duration=2000)]
seq += natives.RX90()
seq.append((acquisition, Delay(duration=natives.RX90().duration)))
seq.append((drive, delays[0]))
seq.append((acquisition, delays[1]))
seq += natives.RX90()
seq.append((acquisition, Delay(duration=natives.RX90().duration)))
seq += natives.MZ()

# introduce sweeper
delay_range = np.arange(100,3000,100)
sweeper = Sweeper(
    parameter = Parameter.duration,
    pulses = delays,
    values=delay_range
)
```

```python
results_t2 = platform.execute([seq], [[sweeper]], nshots=2000, averaging_mode=AveragingMode.CYCLIC)
readout_id_t2 = seq.acquisitions[0][1].id
p1s_t2 = results_t2[readout_id_t2]
```

```python
plot(x=delay_range, y=p1s_t2)
```

```python
popt = fit_exp(delay_range, p1s_t2)
plot_with_fit(delay_range, p1s_t2, fun=exp_decay, popt=popt)
```

```python
print(f"The estimated T2 is {np.round(popt[2])}")
print(f"The simulation value used for T2 is {platform.config("hamiltonian").single_qubit[0].t2[0,1]}")
```

```python
natives = platform.natives.single_qubit[0]

seq = PulseSequence()

# we need to retrieve both the acquisition and the drive channels
acquisition = platform.qubits[0].acquisition
drive = platform.qubits[0].drive
delays = 4* [Delay(duration=2000)]
seq += natives.RX90()
seq.append((acquisition, Delay(duration=natives.RX90().duration)))
seq.append((drive, delays[0]))
seq.append((acquisition, delays[1]))
seq += natives.RX()
seq.append((acquisition, Delay(duration=natives.RX().duration)))
seq.append((drive, delays[2]))
seq.append((acquisition, delays[3]))
seq += natives.RX90()
seq.append((acquisition, Delay(duration=natives.RX90().duration)))
seq += natives.MZ()

# introduce sweeper
delay_range = np.arange(100,3000,100)
sweeper = Sweeper(
    parameter = Parameter.duration,
    pulses = delays,
    values=delay_range / 2
)
```

```python
results_t2_echo = platform.execute([seq], [[sweeper]], nshots=2000, averaging_mode=AveragingMode.CYCLIC)
readout_id_t2_echo = seq.acquisitions[0][1].id
p0s_t2_echo = 1 - results_t2_echo[readout_id_t2_echo]
```

```python
plot(delay_range, p0s_t2_echo, ylabel=r"$P_0$")
```

```python
popt = fit_exp(x=delay_range, y=p0s_t2_echo)
plot_with_fit(x=delay_range, y=p0s_t2_echo, ylabel=r"$P_0$", popt=popt)
```

```python
print(f"The estimated T2 echo is {np.round(popt[2])}")
print(f"The simulation value used for T2 echo is {platform.config("hamiltonian").single_qubit[0].t2[0,1]}")
```
